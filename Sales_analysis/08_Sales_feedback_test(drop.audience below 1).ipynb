### 1. 라벨 인코딩

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
movie['Rate'] = labelencoder.fit_transform(movie['Rate'])
movie['genre_encoded'] = labelencoder.fit_transform(movie['Genre'])
```

```python
dele = movie[movie['Sales'] < 1].index
movie.drop(dele, inplace=True)
movie
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from matplotlib import font_manager
f_path = "C:\Windows\Fonts\malgun.ttf"
font_manager.FontProperties(fname=f_path).get_name()
from matplotlib import rc
rc('font', family='Malgun Gothic')

import warnings
warnings.filterwarnings('ignore')
```

```python
movie.info()
```

```python
numerics = ['int64', 'float64', 'int32']
movie_df = movie.select_dtypes(include=numerics)
```

```python
X = movie_df.drop(['Sales', 'Audience'], axis=1)
y = movie_df['Sales']
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
pred_tr = reg.predict(X_train)
pred_test = reg.predict(X_test)
rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))
rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

print('RMSE of Train Data : ', rmse_tr)
print('RMSE of Test Data : ', rmse_test)
```

```python
sns.set_style('darkgrid')
plt.scatter(y_test, pred_test)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.plot([0,50000], [0,50000], 'r');
```

### 3. 원핫인코딩 + 아웃라이어 제거

```python
movie = pd.read_csv('./datas/final_movie_Rev6.csv', index_col=[0])
movie.reset_index(drop=True, inplace=True)
movie
```

```python
dele = movie[movie['Sales'] < 1].index
movie.drop(dele, inplace=True)
movie
```

```python
fraud = movie['Sales']
quantile_25 = np.percentile(fraud.values, 25)
quantile_75 = np.percentile(fraud.values, 75)

iqr = quantile_75 - quantile_25
iqr_weight = iqr * 1.5
lowest_val = quantile_25 - iqr_weight
highest_val = quantile_75 + iqr_weight

outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
outlier_index, len(outlier_index)
```

```python
movie.drop(outlier_index, axis=0, inplace=True)
```

```python
numerics = ['int64', 'float64', 'uint8']
movie_df = movie.select_dtypes(include=numerics)

X = movie_df.drop(['Sales', 'Audience'], axis=1)
y = movie_df['Sales']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
pred_tr = reg.predict(X_train)
pred_test = reg.predict(X_test)
rmse_tr = (np.sqrt(mean_squared_error(y_train, pred_tr)))
rmse_test = (np.sqrt(mean_squared_error(y_test, pred_test)))

print('RMSE of Train Data : ', rmse_tr)
print('RMSE of Test Data : ', rmse_test)
```

```python
sns.set_style('darkgrid')
plt.scatter(y_test, pred_test)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.plot([0,50000], [0,50000], 'r');
```
